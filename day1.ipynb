# Arithmetic operations

```python
n1 = 10
n2 = 4

print(  f"The addition of {n1} and {n2} is {n1 + n2}"   )
print(  f"{n2} subtracted from {n1} gives {n1 - n2}"    )
print(  f"{n1} multiplied by {n2} gives {n1 * n2}"      )
print(  f"{n1} divided by {n2} gives {n1 /  n2}"        ) #true division
print(  f"{n1} divided by {n2} gives quotient {n1 //  n2}"  ) #floor division
print(  f"{n1} divided by {n2} gives remainder {n1 %  n2}"   ) #remainder
print(  f"{n1} raised to the power of {n2} gives {n1 **  n2}"  ) #rasied to or exponent


```

# filtering & grouping operations

#### rows with ApplicantIncome over 5000

```python
condition = (    df['ApplicantIncome'] > 5000      )

df.loc[        condition     ,     ['Gender', 'ApplicantIncome', 'Loan_Status']              ]
```

#### rows with ApplicantIncome over 5000 & Gender as '0'

```python
condition = (    df['ApplicantIncome'] > 5000      )  &  (     df['Gender'] == 0     )

df.loc[        condition     ,     ['Gender', 'ApplicantIncome', 'Loan_Status']       ]
```

#### rows with ApplicantIncome between 5000-6000 OR CoapplicantIncome between 5000-6000 Gender as '0'

```python
condition1 = (df['ApplicantIncome'].between(5000,6000))

condition2= (df['CoapplicantIncome'].between(5000,6000))

df.loc[      condition1 | condition2 ,           ]
```

# grouping of data

### Loan Status

```python
key =  ['Loan_Status'] 
cols = ['CoapplicantIncome', 'ApplicantIncome', 'LoanAmount'] 
ops = ['min', "max", "mean"]

df.groupby(    key    )   [     cols      ].agg(      ops     )
```

```python
key =  ['Self_Employed'] 
cols = ['CoapplicantIncome', 'ApplicantIncome', 'LoanAmount'] 
ops = ['min', "max", "mean"]

result = df.groupby(    key    )   [     cols      ].agg(      ops     )

result
```

```python
result.columns
```

```python
new_cols = [        entry[0]+"_"+entry[1]    for entry in result.columns                 ]

result.columns = new_cols
```
