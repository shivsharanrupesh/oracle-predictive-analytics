```python
def  addition (    n1 : int or float ,   n2 : int or float   ) -> None :
    #if n1 is either float or int AND n2 is also float or int:
    if(  isinstance(n1,  (int, float)  )  and  isinstance(n2, (int,float) ) ):
        print(f"The addition of {n1} and {n2} is {n1 + n2}")
    # else, print an error message
    else:
        print("incompatible types")
```

```python
addition(10, 20)
addition(400, 20)
addition(100, 200)
addition("harshit", "shukla") #concatenated strings
addition([1,2,3], [2,60,70]) #new list by merging into one extended list
```

### lambda functions


lambda     <input-variables>    :     <output-expression>



```python
#ans is a list of mapping square function on each value of this list
ans = list ( map(   lambda  number :  number ** 2  ,   [1,2,3,4,5]  ) )  #apply function to each item in data collections
ans
```

```python
ans = list ( map(   lambda  number :  number ** 3 ,   [1,2,3,4,5]  ) )  #apply function to each item in data collections
ans
```

```python
ans = list ( map(   lambda  number :  0.9 * number ,   [1,2,3,4,5]  ) )  #apply function to each item in data collections
ans
```

```python
import math

ans = list ( map(   math.factorial  ,   [1,2,3,4,5]  ) )  #apply function to each item in data collections
ans
```

```python
ans = list ( map(   math.sqrt  ,   [1,2,3,4,5]  ) )  #apply function to each item in data collections
ans
```
